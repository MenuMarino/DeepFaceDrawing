```python
from matplotlib.offsetbox import AnnotationBbox, OffsetImage
import matplotlib.pyplot as plt
import numpy as np

disturbance_level = np.array([80, 100, 120])
FID = np.array([76.17, 66.06, 64.6])

fig, ax = plt.subplots()
ax.plot(disturbance_level, FID, marker='*', color='blue')
ax.set_ylim(min(FID) - 5, max(FID) + 10)

img1 = plt.imread('person-face-sketches/test/generated-w1-images/20.jpg')
img2 = plt.imread('person-face-sketches/test/generated-w2-images/20.jpg')
img3 = plt.imread('person-face-sketches/test/generated-w3-images/20.jpg')

imagebox1 = OffsetImage(img1, zoom=0.2)
imagebox2 = OffsetImage(img2, zoom=0.2)
imagebox3 = OffsetImage(img3, zoom=0.2)

ab1 = AnnotationBbox(imagebox1, (disturbance_level[0] + 1, max(FID) + 5), frameon=False, box_alignment=(0.5, 0))
ab2 = AnnotationBbox(imagebox2, (disturbance_level[1] + 1, max(FID) + 5), frameon=False, box_alignment=(0.5, 0))
ab3 = AnnotationBbox(imagebox3, (disturbance_level[2] + 1, max(FID) + 5), frameon=False, box_alignment=(0.5, 0))

ax.add_artist(ab1)
ax.add_artist(ab2)
ax.add_artist(ab3)

ax.set_xlabel('pocas')
ax.set_ylabel('FID')

plt.show()
```
